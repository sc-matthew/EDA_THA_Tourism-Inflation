```python
url_2013 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2013.xlsx'
url_2014 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2014.xlsx'
url_2015 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2015.xlsx'
url_2016 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2016.xlsx'
url_2017 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2017.xlsx'
url_2018 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2018.xlsx'
url_2019 = 'https://github.com/sc-matthew/EDA_THA_Tourism-Inflation/raw/main/Tourist_ArrivalData_2019.xlsx'
```

```python
continent_dict = {
    'East Asia': ['East Asia', 'Asean', 'Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Vietnam', 'China', 'Hong Kong', 'Japan', 'Korea', 'Taiwan', 'Others in East Asia'],
    'Europe': ['Europe', 'Austria', 'Belgium', 'Denmark', 'Finland', 'France', 'Germany', 'Italy', 'Netherlands', 'Norway', 'Russia', 'Spain', 'Sweden', 'Switzerland', 'United Kingdom', 'East Europe', 'Others in Europe'],
    'The Americas': ['The Americas', 'Argentina', 'Brazil', 'Canada', 'USA', 'Others  in America'],
    'South Asia': ['South Asia', 'Bangladesh', 'India', 'Nepal', 'Pakistan', 'Sri Lanka', 'Others in South Asia'],
    'Oceania': ['Oceania', 'Australia', 'New Zealand', 'Others  in Ociania'],
    'Middle East': ['Middle East', 'Egypt', 'Israel', 'Kuwait', 'Saudi Arabia', 'U.A.E.', 'Others in Middle East'],
    'Africa': ['Africa', 'South  Africa', 'Others in Africa'],
}
```

```python
def retrieve_continent(data):
    for continent in continent_dict.items():
        if data in continent[1]:
            return continent[0]

def create_continet_list(dataframe):
    continent_list = []
    for index, row in dataframe.iterrows():
        continent_list.append(retrieve_continent(row['of Residence']))
    return continent_list
```

```python
file_list = [url_2013, url_2014, url_2015,
             url_2016, url_2017, url_2018, url_2019]
df_list = []

for file in file_list:
    temp_df = pd.read_excel(file, skiprows=5, skipfooter=2)
    temp_df.insert(1, "Year", file[-9:-5])
    temp_df['Continents'] = create_continet_list(temp_df)

    if file[-9:-5] in ['2013', '2014', '2015']:
        temp_df.drop(temp_df.columns[7], axis=1, inplace=True)
        df_list.append(temp_df)
    else:
        temp_df.drop(temp_df.iloc[:, 7:22], axis=1, inplace=True)
        df_list.append(temp_df)

df = pd.concat(df_list)
df = df.drop(columns='$US/Day').reset_index(drop=True)
df = df.iloc[:, [6, 0, 1, 2, 3, 4, 5]]
```

```python
df = df.rename(columns={
    'of Residence': 'Country',
    'Unnamed: 1': 'No. Arrival',
    '(Days)': 'Avg. Length of Stay',
    'Baht/Day': 'Spending per capita per day (Baht)',
    'Mil. Baht': 'Tourism Receipts (Mil. Baht)'
})
```

```python
df.drop(['Avg. Length of Stay'], axis=1, inplace=True)
```

```python
delete_list = ['East Asia', 'Europe', 'The Americas',
               'South Asia', 'Oceania', 'Middle East', 'Africa', 'Asean']

for element in delete_list:
    df = df[df['Country'] != element]
```

```python
df_with2013 = df.copy()
df = df[df['Year'] != '2013'].reset_index(drop=True)
```

## Interactive Graph


```python
continent_list = []
country_list = []
year_list = []
dimension_list = []

for element in df.columns:
    if element == 'Continents':
        continent_list = list(df[element].unique())
    elif element == 'Country':
        country_list = sorted(list(df[element].unique()))
    elif element == 'Year':
        year_list = list(df[element].unique())
    else:
        dimension_list.append(element)
```

```python
@interact(Continent=continent_list, Year=year_list, Dimension=dimension_list)
def plot_chart(Continent, Year, Dimension):
    graph = df[(df['Continents'].isin([Continent])) & df['Year'].isin([Year])]
    graph = graph.sort_values(by=Dimension, ascending=False)
    plt.figure(figsize=(24, 6))
    sns.barplot(data=graph, x='Country', y=Dimension, hue='Continents')
    plt.grid(alpha=0.2)

    # Get the current axis
    ax = plt.gca()

    if Dimension == 'No. Arrival':
        plt.title(
            f'Numbers of international tourists from {Continent} by Country in {Year}')
    elif Dimension == 'Tourism Receipts (Mil. Baht)':
        plt.title(
            f"Total Tourism Receipts from {Continent}'s tourists by Country in {Year}")
    else:
        plt.title(
            f"Average {Continent}'s tourists spending per capita per day by Country in {Year}")

    # Define a custom formatter function
    def y_axis_numformat(x, pos):
        return "{:,.0f}".format(x)

    # Apply the formatter to the y-axis
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(y_axis_numformat))

    plt.show()
```

```python
@interact(Country=country_list,  Dimension=dimension_list)
def plot_chart(Country='China', Dimension='No. Arrival'):
    graph = df[(df['Country'].isin([Country]))]
    plt.figure(figsize=(13, 6))
    sns.barplot(data=graph, x='Year', y=Dimension, hue='Country')
    plt.grid(alpha=0.2)

    if Dimension == 'No. Arrival':
        plt.title(
            f'Numbers of international tourists from {Country} entering Thailand from 2014-2019')
    elif Dimension == 'Tourism Receipts (Mil. Baht)':
        plt.title(
            f"Total Tourism Receipts from {Country}'s tourists from 2014-2019")
    else:
        plt.title(
            f"Average {Country}'s tourists spending per capita per day from 2014-2019")
    # Get the current axis
    ax = plt.gca()

    # Define a custom formatter function
    def y_axis_numformat(x, pos):
        return "{:,.0f}".format(x)

    # Apply the formatter to the y-axis
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(y_axis_numformat))

    plt.show()
```

```python
url_list = [th_inflation, vn_inflation]
df_list = []

for file in url_list:
    raw = pd.read_csv(file, skiprows=1)
    raw.drop(raw.iloc[:, 2:5], axis=1, inplace=True)
    temp1 = raw.drop(raw.index[0:2]).astype('str')
    temp1 = temp1.drop(raw.index[8:])

    if 'Vietnam' in file:
        temp1['Vietnam'] = pd.to_numeric(
            temp1['Vietnam'].str.strip('%'), errors='coerce')
        temp1 = temp1.rename(columns={
            'Vietnam': 'Vietnam Inflation (% Change)'})
        df_list.append(temp1)
    else:
        temp1['Thailand'] = pd.to_numeric(
            temp1['Thailand'].str.strip('%'), errors='coerce')
        temp1 = temp1.rename(columns={
            'Thailand': 'Thailand Inflation (% Change)'})
        df_list.append(temp1)


df_merged = pd.merge(df_list[0], df_list[1], on='Year')
df_merged = df_merged.sort_values(
    by=['Year']).reset_index().drop(columns=['index'])
```
